```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
```

```python
#read from train.csv
td = pd.read_csv('train.csv')
```

```python
#no. of rows and coloumn
td.shape
```

```python
#display elements of data (default value = 5)
td.head()
```

```python
td.info() #non null means values which are present
```

```python
td.isnull().sum()
```

```python
#delete column cabin as it has max null values
td = td.drop(columns = 'Cabin', axis = 1)
```

```python
#fillna = fill all missing values that are na(not available)
td['Age'].fillna(td['Age'].mean(),inplace = True)
td['Fare'].fillna(td['Fare'].mean(),inplace = True)
td['Embarked'].fillna(td['Embarked'].mode()[0],inplace = True)
td.isnull().sum()
```

```python
td.replace({'Sex':{'male':0, 'female':1}, 'Embarked':{'S': 0, 'C' : 1, 'Q' : 2}}, inplace = True)
```

```python
td.head()
```

```python
x = td.drop(columns = ['PassengerId', 'Name', 'Ticket', 'Survived'], axis = 1)
y = td['Survived']
print(x)
print(y)
```

```python

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
# model training(logistic regression)
```

```python
#loading the logistic regression from inbuilt libraries i.e sklearn
model = LogisticRegression()
```

```python
#training the logistic regression model with training data 
```

```python
model.fit(x_train, y_train)
```

```python
#model evaluation
```

```python
#accuracy score for train data
x_train_prediction = model.predict(x_train)
print(x_train_prediction)
```

```python
training_data_accuracy = accuracy_score(y_train, x_train_prediction)
print('Accuracy score of training data is : ', training_data_accuracy)
```

```python
#accuracy score on test data (imp)
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(y_test, x_test_prediction)
print('Accuracy score of test data is :', test_data_accuracy)
```
